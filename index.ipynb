## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
import seaborn as sns


# List of features to plot against SalePrice
features = ['LotArea', '1stFlrSF', 'GrLivArea']

# Create scatter plots
plt.figure(figsize=(15, 5))

for i, feature in enumerate(features):
    plt.subplot(1, 3, i + 1)
    sns.scatterplot(data=ames_subset, x=feature, y='SalePrice', alpha=0.6)
    plt.title(f'{feature} vs SalePrice')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')

plt.tight_layout()
plt.show()
````

# Your written answer here - do these seem like good candidates for linear regression?
For the GrLivArea and 1stFlrSF, Yes. since they  show some degrees of suitability for linear regression:

GrLivArea appears to be a strong candidate. The scatter plot shows a clear positive linear relationship with SalePrice, with most data points clustered along an upward trend. 
Despite a few outliers, this feature should perform well in a linear regression model.

1stFlrSF also shows a positive trend with SalePrice, though the relationship is slightly more dispersed than GrLivArea. 
It still appears to be a reasonable predictor and is likely to contribute positively to a multivariable model.

However, LotArea, shows a much weaker relationship. The data is highly skewed and includes some extreme outliers, which may reduce its effectiveness in a linear regression context. 
It may benefit from preprocessing, such as log transformation or outlier removal, before including it in a model.

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
X = ames_subset['GrLivArea']
y = ames_subset['SalePrice']
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

# Add a constant to X for the intercept term
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())
````

**Your written answer here - interpret model results**

R-squared: 0.502 - About 50.2% of the variation in SalePrice is explained by GrLivArea alone. That’s a moderate fit for a single-variable model — quite decent.

const - has a coefficient of 18,570.44 - that is the  baseline sale price when GrLivArea is 0 
GrLivArea - has a coefficient of 107.13. Which means tha for every additional square foot of above-ground living space, the predicted sale price increases by $107.13, on average.

F-statistic: 1471 with p-value (Prob F-statistic): 4.52e-223
Extremely significant. This tells us that the overall model is statistically valid.

The p-values for both variables are 0.000, meaning both coefficients are statistically significant (p < 0.05).

The 95% confidence interval for GrLivArea is [101.65, 112.61], indicating we’re 95% confident that the true effect lies in that range.




## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X

X = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
````

````python
# Your code here - fit model and display results

# Fit the model
multi_model = sm.OLS(y, X).fit()

# Show the model summary
print(multi_model.summary())
````

# Your written answer here - interpret model results. Does this model seem better than the previous one?

R-squared improvement: The multiple model explains 6.3% more variance in SalePrice than the simple model — a moderate improvement.

Lower AIC in the multiple model also indicates a better fit (lower is better).

By including LotArea and 1stFlrSF, we capture more variation in sale prices.

This model is better than the previous one:

It explains more variation in sale prices.

It incorporates more features that are relevant.

It provides more detailed insights into what drives home values.



## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor

#import statsmodels.api as sm
#import matplotlib.pyplot as plt

# Define dependent and independent variables
X = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
y = ames_subset['SalePrice']

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Plot partial regression plots
fig = sm.graphics.plot_partregress_grid(model, fig=plt.figure(figsize=(12, 8)))
plt.tight_layout()
plt.show()
````

# Your written answer here - explain what you see, and how this relates
# to what you saw in Step 1. What do you notice?

Among the predictors, GrLivArea has the strongest unique linear relationship with SalePrice after accounting for the effects of other predictors in the model. 
This is indicated by the steepest slope and the relatively tighter scatter in its partial regression plot.

1stFlrSF also shows a positive and noticeable unique relationship with SalePrice, although not as strong as GrLivArea.

LotArea has a weaker unique positive relationship with SalePrice.

*How it relates to step one output*

The direction of the relationships (positive in all cases) is generally consistent between the raw scatter plots and the partial regression plots.

The relative strength of the relationships observed in the raw scatter plots (strongest for GrLivArea, moderate for 1stFlrSF, weakest for LotArea) is also reflected in the slopes and scatter of the corresponding partial regression plots.

The partial regression plots provide a more nuanced view by isolating the unique linear contribution of each predictor. While the raw scatter plots show the overall association, the partial plots reveal the impact of each predictor holding the others constant. 

The fact that GrLivArea still shows a strong positive relationship in its partial plot highlights its independent importance in predicting SalePrice.


## Level Up (Optional)

Re-create this model in scikit-learn, and check if you get the same R-Squared and coefficients.

````python
# Your code here - import linear regression from scikit-learn and create and fit model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Define features and target
X = ames_subset[['LotArea', '1stFlrSF', 'GrLivArea']]
y = ames_subset['SalePrice']

# Create and fit the model
model_sklearn = LinearRegression()
model_sklearn.fit(X, y)

# Predict using the model
y_pred = model_sklearn.predict(X)
````

````python
# Your code here - compare R-Squared
# R-squared
r2 = r2_score(y, y_pred)

# Print R-squared and coefficients
print("R-squared:", r2)
````

````python
# Your code here - compare intercept and coefficients

print("Intercept:", model_sklearn.intercept_)

# Map coefficients to feature names
coeff_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model_sklearn.coef_
})
print(coeff_df)
````
